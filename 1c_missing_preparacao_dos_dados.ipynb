# Dados missing


## Teoria

    Explícitos
    - Marcados explicitamente com NULL ou NA (No Pandas: NaN)
    
    Implícito:
    – Podem não estar lá (observação ausente)
    – Uso de placeholders: 99, -1, 1900...

    Por que tratar?
    – Podem introduzir viés na análise.

    Como identificar?
    – Para detectar usar funções de resumo como .info(), .describe() ou próprias como .isna(), .isnull(), .notna().
    - Utilizar a biblioteca missingno para diagnosticar o tipo de dado perdido

## Prática

#### Usando Missingno

Podemos diagnosticar o tipo de dado perdido com a biblioteca missingno. Ela é essencial para nos mostrar o tamanho do estrago, ou seja, conseguimos visualmente identificar possíveis padrões que justifiquem a ausência de dados e classifiquem eles como não aleatório, aleatório ou completamente ao acaso.

```python
msno.matrix(dados)
```

```python
msno.bar(dados)
```

```python
msno.heatmap(dados)
```

```python
msno.dendrogram(dados)
```

## Tratamentos 

    Para tratar, podemos:
    – Remover a observação completamente.
    – Usar métodos de imputação de valores:
        • Estatística de resumo: média, mediana ou moda.
        • Criar variável para sinalizar coluna ausente.
        • Substituir por outlier


```python
#contagem de dados perdidos
dados.isnull().sum()
```

```python
# exemplo de substituição por outlier ou missing implicito
dados.fillna(-3, inplace=True)
```
